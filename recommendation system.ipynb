```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('DB_new.csv',encoding='CP949') # 공연 예정, 공연 중인 DB 변수에 저장
data.tail() 
```

```python
data_old = pd.read_csv('DB_old.csv',encoding='utf-8') # 지난 공연 DB 변수에 저장
data_old.head()
```

```python
# 사용자가 시청한 공연 목록을 data에 추가 
usr_df = data_old[data_old['title'] == '국카스텐 콘서트 HAPPENING 〈ESCAPE VELOCITY〉']
data = data.append(usr_df,ignore_index=True)

data.tail()
```

```python
# 데이터 선호도에 따른 필터링을 위한 전처리
num = len(data)

data_genres= data['genres']
genres_list = data_genres[0].split('|')

for i in range (num):
    li2 = data_genres[i].split('|')
    for j in li2:
        if j not in genres_list:
            genres_list.append(j)
            
data_actor= data['actor']
actor_list = data_actor[0].split('|')

for i in range (num):
    li3=data_actor[i].split('|')
    for j in li3:
        if j not in actor_list:
            actor_list.append(j)
```

```python
for i in range(len(genres_list)):
    data['genres_'+genres_list[i]]=0
    
for i in range(len(actor_list)):
    data['actor_'+actor_list[i]]=0


for i in range (num):
    li2=data_genres[i].split('|')
    for j in li2:
        data.loc[i,['genres_'+j]]=1

for i in range (num):
    li3=data_actor[i].split('|')
    for j in li3:
        data.loc[i,['actor_'+j]]=1
```

```python
data.head()
```

```python
# 사용자의 원하는 가격대 설정
user_price_min = int(input("Min(0~): "))  #최저 가격
user_price_max = int(input("Max(~300000): "))  #최대 가격

num = len(data) - 1

for i in range(num):  # 설정한 가격대를 벗어난 항목 필터링 
    if data['price'][i] > user_price_max or data['price'][i]< user_price_min:
        data = data.drop(i)
```

```python
# 원하지 않는 장르 제외

user_dislike=input("Dislike genres: ")
dislike="genres_"+user_dislike
if user_dislike == 'none' :
    print("No input")
else:
    data_index=data[data[dislike]==1].index
    data=data.drop(data_index)

data = data.reset_index(drop=True)
```

```python
data # 장르 및 가격대의 필터링이 잘 이루어졌는지 확인 
```

```python
# 선호하는 장르에 가중치 부여
num= len(data) - 1

user_genres = input("Like Genres: ")

data_keyword = data['genres']+'|'+data['actor'] 
if user_genres == 'none' :
    print("No input")
else:
    data_genres_input= data['genres_'+user_genres]
    for i in range(num):
        if data_genres_input[i] == 1:
            data_keyword[i]+=('|'+user_genres)*3
```

```python
# 선호하는 출연자에 가중치 부여
user_actor = input("Actor: ")

if user_actor == 'none' :
    print("No input")
else:
    data_actor_input= data['actor_'+user_actor]
    for i in range(num):
        if data_actor_input[i] == 1:
            data_keyword[i]+=('|'+user_actor)*3
```

```python
# 선호하는 장르가 기존 데이터에 없을 경우 추가 
flag = 1
li = data_keyword[num].split('|')

for i in li:
    if user_genres == 'none' or user_genres == i:
        flag = 0
        break

if flag ==1:
    data_keyword[num]= data_keyword[num]+"|"+user_genres

data_keyword[num]
```

```python
# 선호하는 출연자가 기존 데이터에 없을 경우 추가 
flag = 1
li = data_keyword[num].split('|')

for i in li:
    if user_actor == 'none' or user_actor == i :
        flag = 0
        break
        
if flag ==1:
    data_keyword[num]= data_keyword[num]+"|"+user_actor

data_keyword[num]
```

```python
# 데이터에 있는 문자 '|'를 ' '(공백)으로 대체
num = num+1
for i in range(num):
    data_keyword[i] = data_keyword[i].replace("|"," ")
    data['genres'][i]=data['genres'][i].replace("|"," ")
    data['actor'][i]=data['actor'][i].replace("|"," ")
```

```python
#단어 벡터화
from sklearn.feature_extraction.text import CountVectorizer

count_vector = CountVectorizer(analyzer='word', ngram_range=(1, 1))
```

```python
#코사인 유사도 적용
from sklearn.metrics.pairwise import cosine_similarity

count_matrix = count_vector.fit_transform(data_keyword)
cosine_sim = cosine_similarity(count_matrix, count_matrix).argsort()[:, ::-1]
```

```python
#추천 알고리즘
def recommendation(db, ontect_title, top=10):
    target_index = db[db['title'] == ontect_title].index.values
    sim_index = cosine_sim[target_index, :top].reshape(-1)
    sim_index = sim_index[sim_index != target_index][:8]
    sim_data=db.iloc[sim_index]

    return sim_data
```

```python
#추천 공연 추출
recommendation(data, ontect_title='국카스텐 콘서트 HAPPENING 〈ESCAPE VELOCITY〉')
```
